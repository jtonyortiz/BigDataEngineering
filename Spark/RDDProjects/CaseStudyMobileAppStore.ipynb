```python
import pandas as pd
import numpy as np
from pyspark import SparkContext
sc = SparkContext()
```

```python
file = "AppleStore.csv"
filedata = sc.textFile(file)
```

```python
rows = filedata.map(lambda line: line.split(","))
```

```python
keyword = rows.keyBy(lambda word: word.upper()[0])
```

```python
df = pd.read_csv(file)
```

```python
data_top = df.head()
print(data_top)
```

```python
byte_list = df["size_bytes"].tolist()
```

```python
byte_list
```

```python
def convert_to_MB(x):
    return x / 1048576
```

```python
MB_list = []
for x in byte_list:
    MB_list.append(convert_to_MB(x))

MB_list
```

```python
def convert_to_GB(x):
    return x / 1048
```

```python
GB_list = []
for x in MB_list:
    GB_list.append(convert_to_GB(x))
GB_list
```

```python
df["size_MB"] = MB_list #Adding MB size to dataframe
df["size_GB"] = GB_list #Adding Gb size to dataframe
```
