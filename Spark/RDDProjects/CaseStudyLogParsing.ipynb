```python
from pyspark.sql import SparkSession
spark = SparkSession \
    .builder\
    .master("yarn")\
    .appName("RDD Creation")\
    .getOrCreate()
```

```python
spark
```

```python
inputRDD = spark.sparkContext.textFile("access.log")
```

### Finding out how many 404 HTTP Codes are in Access Logs

```python
RDD_404_ = inputRDD.filter(lambda x: "404" in x)
```

```python
RDD_404_.count()
```
