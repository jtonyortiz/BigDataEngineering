```python
import pandas as pd
import numpy as np
```

```python
ds_score = pd.read_csv('DSScoreTerm1.csv')
math_score = pd.read_csv('MathScoreTerm1.csv')
physics_score = pd.read_csv('PhysicsScoreTerm1.csv')
```

```python
#Remove the name and ethnicity column for each DataFrame
ds_score.drop(['Name', 'Ethinicity'], axis=1)
math_score.drop(['Name', 'Ethinicity'], axis=1)
physics_score.drop(['Name', 'Ethinicity'], axis=1)
```

```python
#Fill Scores with Zeros:
ds_score['Score'] = ds_score['Score'].fillna(0)
math_score['Score'] = ds_score['Score'].fillna(0)
physics_score['Score'] = ds_score['Score'].fillna(0)
```

```python
#Merge the dataframes into one:
merged_df = pd.concat([ds_score, math_score, physics_score], ignore_index=True)
```

```python
merged_df
```

```python
#Change Sex(M/F) Column to 1/2 for further analysis
merged_df = merged_df.replace('M', 1)
```

```python
merged_df = merged_df.replace('F', 2)
```

```python
#Saving DataFrame to CSV file:
merged_df.to_csv('ScoreFinal.csv')
```
