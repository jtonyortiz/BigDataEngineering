## Assignment II Part 3 - Manufacturing

```python
manu = pd.read_csv('Cars2015.csv')
manu
```

```python


df = pd.read_csv('Cars2015.csv')
df = df.drop(['Model','Type','LowPrice','HighPrice','Drive','CityMPG','HwyMPG','FuelCap','Length'], axis=1)
df = df.drop(['Width','Wheelbase','Height','UTurn','Weight','Acc030','Acc060'],axis=1)
df = df.drop(['QtrMile','PageNum','Size'],axis=1)
df['count'] = df.groupby('Make')['Make'].transform('count')
df.drop_duplicates('Make',inplace=True)
df.sort_values(by='count',ascending=False,inplace=True)
plt.figure(figsize=(50,8))
ax1 = plt.subplot(121, aspect='equal')
df.plot(kind='pie', y = 'count', ax=ax1, autopct='%1.1f%%',startangle=90, shadow=False, labels=df['Make'], legend = False, fontsize=10) 
plt.show()
```

## Assignment II Part 4

```python
info_df = pd.read_csv('Info.csv') #Contains CSV data from Assignment 
info_df.head()
```

```python
#Describe the unitprice:
info_df.info()
```

```python
# Create anew DataFramehaving columnsasname,net_price,date,and group all the records according to name.
new_df = info_df[['name', 'net_price', 'date']].filter(['name', 'net_price', 'date']).groupby('name', as_index=False).sum()
new_df
```

```python
x = new_df['name']
y = new_df['net_price']
plt.figure(figsize=(20, 7))
plt.xticks(ticks=list(range(20)), labels=x, rotation = 'vertical',  size=14, color='red')# upper-case
plt.xlabel('Name', size=20, color='steelblue', weight='bold')
plt.ylabel('Total Sales', size=20, color='steelblue', weight='bold')
plt.title('Total Sales by Each Customer', size=20)
plt.bar(x, y)
```
