## Merging & Joins 
    - Inner: All of the common keys
    - Outer: All of the rows
    - Left: All of the ids of the left data
    - Right: All of the ids of the right data

```python
data1 = pd.DataFrame({'Enrol_NO': ['A11', 'A12', 'A13', 'A14', 'A15'], 
                     'Names': ['A', 'B', 'C', 'D', 'E'],
                     'Age':[27, 28, 26, 27, 29],
                     'Height': [160, 165, 155, 167, 164]})
```

```python
data2 = pd.DataFrame({'Enrol_NO':['A12', 'A14', 'A16', 'A17', 'A11'],
                     'Diet':['V', 'NV', 'V', 'V', 'NV']})
```

```python
report = pd.merge(data1, data2, how='inner')
report
```

```python
report = pd.merge(data1, data2, how='outer')
report
```

```python
report = pd.merge(data1, data2, how='left')
report
```

```python
report = pd.merge(data1, data2, how='right')
report
```

```python
pd.merge(data1, data2, on='Enrol_NO')
```
