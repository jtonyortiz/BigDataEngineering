```python
plt.figure(figsize = (10,5))
plt.bar(freq_ser.index, freq_ser.values, color = ['mediumorchid', 'lightcoral', 'paleturquoise',
                                                  'olive', 'lightsalmon', 'lightseagreen'] , edgecolor = 'grey')
plt.xlabel('\nTypes of Cars', size = 20, color = 'steelblue', weight = 'bold')
plt.ylabel('Count\n', size = 20, color = 'steelblue', weight = 'bold')
plt.xticks(ticks = list(range(6)),labels = freq_ser.index.str.upper(), rotation = 0, size = 14, color = 'steelblue')
plt.ylim((0,60))
plt.yticks(ticks = range(0,61, 15),size = 14, color = 'steelblue', rotation = 0)
for i in range(len(freq_ser)):
    plt.annotate(freq_ser.values[i], xy = (i, freq_ser.values[i]+2), size = 14)
    
plt.title('Distribution of Types of Cars\n', size = 24, color = 'chocolate', weight = 'bold')
plt.show()
```

## 2. Pie Chart

```python
plt.figure(figsize = (5,5))
freq_ser = cars_data.Type.value_counts()
plt.pie(freq_ser.values, radius = 1.2, labels = freq_ser.index,
        colors = ['mediumorchid', 'lightcoral', 'paleturquoise','olive', 'lightsalmon', 'lightseagreen'],
        autopct= '%.2f %%', shadow = True,
        textprops={'size': 12, 'color':'black'},
        wedgeprops={'edgecolor': 'white', 'width': 0.75}
       )
plt.show()
```

```python
# filter for tof 5 makes 
top5 = cars_data.Make.value_counts()[:5].index
top5
```

```python
sub= cars_data[cars_data.Make.isin(top5)]
```

```python
freq_tab = pd.crosstab(sub.Make, sub.Type)[['SUV', 'Sedan', 'Sporty']]
freq_tab
```

```python
freq_tab.plot.bar()
plt.show()
```

```python
plt.bar(freq_tab.index, freq_tab.SUV, color = 'green')
plt.bar(freq_tab.index, freq_tab.Sedan, color = 'orange', bottom = freq_tab.SUV)
plt.bar(freq_tab.index, freq_tab.Sporty, color = 'blue', bottom = freq_tab.SUV + freq_tab.Sedan)
plt.show()
```
